```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("assign_1.csv")
```

```python
df.describe()
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df= df[df['YearBuilt'].notna()]
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
df['Car'].fillna(df.Car.mode()[0],inplace=True)
df['YearBuilt'] = df['YearBuilt'].fillna(df['YearBuilt'].median())
df['CouncilArea'].fillna(value="new type",inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
df['Type'] = encoder.fit_transform(df['Type'])
```

```python
df.Method.unique()
```

```python
df['Method']=encoder.fit_transform(df['Method'])
```

```python
df.SellerG.unique()
```

```python
df['SellerG'] = encoder.fit_transform(df['SellerG'])
```

```python
df['Regionname'] = encoder.fit_transform(df['Regionname'])
```

```python
df['CouncilArea'] = encoder.fit_transform(df['CouncilArea'])
```
